```python
#dict
fruit_protein = {
    "Avocado": 2.0,      # grams of protein
    "Guava": 2.6,
    "Blackberries": 2.0,
    "Oranges": 0.9,
    "Banana": 1.1,
    "Apples": 0.3,
    "Kiwi": 1.1,
    "Pomegranate": 1.7,
    "Mango": 0.8,
    "Cherries": 1.0
}
```

```python
s2 = pd.Series(fruit_protein,name = "Protein")
s2
```

```python
#Conditional Selection

s2>1   #Greater than 1
```

```python
s2[s2>1]
```

```python
#Logical Operators : AND,OR,NOT

[(s2>0.5) & (s2<2)]
```

```python
s2[(s2>0.5) & (s2<=2)]
```

```python
s2[(s2>0.5) | (s2<=2)]
```

```python
#NOT
s2[~(s2>1)]
```
