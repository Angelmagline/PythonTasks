```python
data = np.array([
    [1, 23, 172.4, 68.2, 88.5],
    [2, 35, 181.2, 82.1, 74.3],
    [3, 29, 165.3, 59.4, 91.2],
    [4, 41, 177.8, 79.6, 67.9],
    [5, 22, 169.0, 63.7, 84.1],
    [6, 33, 174.5, 70.4, 78.8],
    [7, 27, 182.1, 85.0, 72.5],
    [8, 30, 168.9, 62.9, 89.3],
    [9, 25, 176.2, 74.2, 80.7],
    [10, 38, 179.5, 77.1, 69.4]
])


header = "id,age,height_cm,weight_kg,score"
np.savetxt("data.csv", data, delimiter=",", header=header, comments="")

print("CSV file 'data.csv' created successfully!")
```

```python
print(data)
```

```python
print("Loaded Data:\n", data)
```

```python
print("Size (total elements):", data.size)
print("Shape (rows, columns):", data.shape)
print("Dimensions:", data.ndim)
print("Data type:", data.dtype)
```

```python
#Sorting
print("\nSORTING")

# Sort each row
print("Row-wise sort:\n", np.sort(data, axis=1))

# Sort each column
print("\nColumn-wise sort:\n", np.sort(data, axis=0))
```

```python
#Reshaping & Array convertions
flat = data.flatten()
print("\nFlattened data:\n", flat)
```

```python
reshaped = flat.reshape(5, 10)   # Just for demonstration
print("\nReshaped to 5×10:\n", reshaped)
```

```python
print("\nTranspose of reshaped:\n", reshaped.T)
```

```python
print("\nIndexing & Slicing")

print("First row:", data[0])
print("Last row:", data[-1])
```

```python
# Slicing rows
print("First 3 rows:\n", data[:3])
```

```python
# Slicing columns (age = column 1)
print("Age column:", data[:, 1])
```

```python
# Filter
print("People with age > 30:\n", data[data[:, 1] > 30])
```

```python
#Math Operations

print( np.sum(data))
print( np.mean(data))
print( np.mean(data, axis=0))
print( np.std(data))
print( np.min(data))
print( np.max(data))
```
